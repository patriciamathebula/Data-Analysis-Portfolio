# World largest Oil Reserve and Consumption

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("World Crude Oil Reserves from 1995 to 2021.csv")
print(df.head())
```

```python
df.index=df.iloc[:,0]
df=df.T[1:]
print(df.head())
```

```python
df.tail()
```

```python
#Remove unnecessary columns
drop_cols=[]
for i,col in enumerate(df.columns):
    if df.iloc[-1,i]==0:
        drop_cols+=[i]
#print(drop_cols)        
df=df.drop(df.columns[drop_cols], axis=1)
display(df[0:5])
```

```python
df.describe()
```

```python
df2=df.iloc[-1:,0:-1].T
df2=df2.sort_values('2021',ascending=False)
nations=df2.index.tolist()
print(nations)
```

```python
plt.figure(figsize=(12,6))
for column in nations[0:20]:
    plt.plot(df.index, df[column].astype(float), label=column)
plt.xlabel('Year')
plt.ylabel('TWh')
plt.title('Oil Reserves Top 20')
plt.legend()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.xticks(rotation='vertical')
plt.grid(True)
plt.show()
```

From the plot above, we get the top 20 countries with the highest oil reserviors in the world from the year 1995 to 2021. 
It also show how our top 20 contries have been increasing/ decreasing in the reserves for the past 26 years.
